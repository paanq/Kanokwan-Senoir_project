```python
import numpy as np
from scipy.optimize import linprog
from itertools import combinations
from colorama import Fore, Back, Style
```

```python
def all_diag_combination(dim):
    possible_combination = set(combinations([-1, 1] * dim ,dim))
    return list(np.diag(item) for item in possible_combination)
```

```python
dim = int(input("Enter M: "))
din = int(input("Enter N: "))
itr = int(input("Enter Number of iteration: "))
mul = float(input("Enter width of interval: "))


print("Enter Matrix Ac :")
ac = []
for r in range(dim):
    ac_1 = []
    for c in range(din):
        value_a = input("Enter 2 numbers (with a comma in between): ")
        a, b = value_a.split(',')
        a, b = float(a), float(b)
        ac_1.append((a+b)*0.5)
    ac.append(ac_1)
    
print("Enter Matrix Bc :")
bc = []
for r in range(dim):
    value_b = input("Enter 2 numbers (with a comma in between): ")
    c, d = value_b.split(',')
    c, d = float(c), float(d)
    bc.append((c+d)*0.5)
    
print("Enter Matrix Cc :")
cc = []
for r in range(din):
    value_c = input("Enter 2 numbers (with a comma in between): ")
    e, f = value_c.split(',')
    e, f = float(e), float(f)
    cc.append((e+f)*0.5) 
    
print("Ac =", ac)
print("Bc =", bc)
print("Cc =", cc)

Ac = np.array(ac)
Bc = np.array(bc)
Cc = np.array(cc)
res_1 = []
for i in range(itr):
    A = []
    B = []
    Al = Ac - (np.ones((dim,din))*mul*i)
    Au = -(Ac + (np.ones((dim,din))*mul*i))
    Bl = -(Bc - (np.ones((dim))*mul*i))
    Bu = Bc + (np.ones((dim))*mul*i)
    Cl = Cc - (np.ones((din))*mul*i)
    Cu = -(Cc + (np.ones((din))*mul*i))
    A = np.append( arr=Al, values=Au, axis=0)
    B = np.append( arr=Bu, values=Bl, axis=0)
    res = linprog(Cl, A_ub = A, b_ub = B, method = 'revised simplex' )
    #print('i =',i, ': LB = %.4f' %res['fun'])
    for r in range(2**dim):
        Aye = []
        By = []
        aye = Ac - all_diag_combination(dim)[r].dot(np.ones((dim,din))*mul*i)
        by = Bc + all_diag_combination(dim)[r].dot(np.ones((dim))*mul*i)
        Cy = Cc + (np.ones((din))*mul*i)
        Aye = np.append( arr=aye, values=-aye, axis=0)
        By = np.append( arr=by, values=-by, axis=0)
        Res = linprog(Cy, A_ub = Aye, b_ub = By, method = 'revised simplex' )
        res_1.append(Res['fun'])
    print('i =',i,': LB =%.4f' %res['fun'], ', UB =%.4f ' %max(res_1))
```
