```python
import numpy as np 

from scipy.optimize import linprog

from itertools import combinations

from colorama import Fore, Back, Style

import math as m
```

```python
def all_diag_combination(dim): 

    possible_combination = set(combinations([-1, 1] * dim ,dim))

    return list(np.diag(item) for item in possible_combination)
```

```python
dim = int(input("Enter M: ")) 

din = int(input("Enter N: "))

 

################################# input matrix A ################################

 

print("Enter Matrix A :")                           

Ac = []              #center of matrix A

Au = []              #upper bound of matrix A

Al = []               #lower bound of matrix A

delta = []           #radius of matrix A

A = []               #matrix A

for m in range(dim):

    Ac_1 = []

    Au_1 = []
    
    Al_1 = []

    delta_1 = []
    
    a_1 = []

    for n in range(din):
        
        a_2 = []

        value_A = input("Enter 2 numbers (with a comma in between): ")                  #input matrix A in the pattern of (a, b)

        a, b = value_A.split(',')                                                       #splits a and b

        a, b = float(a), float(b)                                                       #transfrom str(a), str(b) to float(a), float(b) respectively

        Ac_1.append((a+b)*0.5)                                                                                      

        Au_1.append(b)                                                                  #add b(upper bound) in upper bound matrix
        
        Al_1.append(a)

        delta_1.append((b-a)*0.5)
        
        a_2.append(a)
        
        a_2.append(b)
        
        a_1.append(a_2)

    Ac.append(Ac_1)

    Au.append(Au_1)
    
    Al.append(Al_1)

    delta.append(delta_1)
    
    A.append(a_1)

print("Matrix A is", A)   

print("Center matrix of A =", Ac)

print("Lower bound matrix of A =", Al)

print("Upper bound matrix of A =", Au)

print("Radius matrix of A =", delta, '\n')

 

################################# input matrix B ################################

 

print("Enter Matrix B :")

Bc = []         #center of matrix b

Bu = []        #upper bound of matrix b

Bl = []         #lower bound of matrix b

tuda = []       #radius of matrix b

B = []          #matrix B

for m in range(dim):
    
    b_1 = []

    value_B = input("Enter 2 numbers (with a comma in between): ")

    c, d = value_B.split(',')

    c, d = float(c), float(d)

    Bc.append((c+d)*0.5)
    
    Bu.append(d)

    Bl.append(c)

    tuda.append((d-c)*0.5)
    
    b_1.append(c)
    
    b_1.append(d)
    
    B.append(b_1)
   
print("Matrix B is", B)
    
print("Center matrix of B =", Bc)

print("Lower bound matrix of B =", Bl)

print("Upper bound matrix of B =", Bu)

print('Radius matrix of B =', tuda, '\n')

 

################################# input matrix C ################################

 

print("Enter Matrix C :")

C = []             #matrix c

for r in range(din):

    value_C = float(input("Enter number: "))

    C.append(value_C)

   

print("Matrix C =", C)
```

```python
bnd = [] 

p_inf = float("inf")

for e in range(din):
    
    bnd.append((0, p_inf))
    
print(bnd)
```

```python
############## find lower bound of the problem ###############
A_low1 = np.append( arr = Al, values = Au, axis = 0)
B_low1 = np.append( arr = Bu, values = Bl, axis = 0)
res_low1 = linprog(C, A_ub = A_low1, b_ub = B_low1, bounds = bnd, method = 'revised simplex')
#print(res_low1, '\n')
if res_low1['status'] == 0:
    print('Lower bound =', res_low1['fun'], '\n', 'Solution of lower bound case is', res_low1['x'])
else:
    print('Lower bound case: ', res_low1['message'], '\n')

############## find upper bound of the problem ###############
status_up1 = []

x1_up1 = []

print('From', '\n')

for q in range(2**(din)):
    
    x2_up1 = []
    
    aye_up1 = Ac - (all_diag_combination(dim)[q]).dot(delta)
    
    aye_up1 = np.array(aye_up1)
    
    by_up1 = Bc + (all_diag_combination(dim)[q]).dot(tuda)
    
    res_up1 = linprog(C, A_eq = aye_up1, b_eq = by_up1, bounds = bnd, method = 'revised simplex')
    
    #print(res_up1, '\n')
    
    status_up1.append(res_up1['status'])
    
    x2_up1.append(res_up1['fun'])
    
    x2_up1.append(res_up1['x'])
    
    print('Case ', q+1)
    
    if 1 == res_up1['status']:

        print('Iteration limit reached', '\n')

    elif 2 == res_up1['status']:

        print('Upper Bound:', 'Problem appears to be infeasible', '\n')

    elif 3 == res_up1['status']:

        print('Upper Bound:', 'Problem appears to be unbounded', '\n')

    elif 4 == res_up1['status']:

        print('Numerical difficulties encountered', '\n')

    elif 5 == res_up1['status']:

        print('No constraints; turn presolve on', '\n')

    elif 6 == res_up1['status']:

        print('Guess x0 cannot be converted to a basic feasible solution', '\n')

    else:
        
        x1_up1.append(x2_up1)
        
        print('Optimal value of this case is', res_up1['fun'], '\n')
        
#print(status_up1)

x1_up1 = sorted(x1_up1,key=lambda l:l[0], reverse=True)         #sort all possible optimal value

status2_up1 = []

if 3 in status_up1:
    
    status2_up1.append(3)
    print("Hence, this problem appears to be unbounded.", '\n')
    
elif 2 in status_up1:
    
    status2_up1.append(2)
    print("Hence, this problem appears to be infeasible.", '\n')
    
else:
    
    status2_up1.append(0)
    print("Hence, this upper bound of this problem is ", x1_up1[0][0], '\n', 'where the solution of upper bound case is x = ', x1_up1[0][1], '\n')
    
if res_low1['status']==0 and 0 in status2_up1:
    print("GG")
else:
    print('Since, this problem seems to be not have any strong solution. Hence, it not has regret.')
```
